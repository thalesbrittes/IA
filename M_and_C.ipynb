```
class State():
    def __init__(self,LadoEsq, LadoDir, bote):
        self.LadoEsq = LadoEsq
        self.LadoDir = LadoDir
        self.bote = bote
        self.pai = None
        self.filhos = []


    def verifica(self):
        if(self.LadoEsq[0] > 0 and self.LadoEsq[1] > self.LadoEsq[0]):
            return False #Missionários na esquerda é maior que 0 e menor que o número de canibais no mesmo lado
        if(self.LadoDir[0] > 0 and self.LadoDir[1] > self.LadoDir[0]):
            return False #Missionários na direita é maior que 0 e menor que o número de canibais no mesmo lado
        if(self.LadoEsq[0] < 0 or self.LadoEsq[1] < 0 or self.LadoDir[0]<0 or self.LadoEsq[1]<0):
            return False #Canibais e missionários em um dos lados não pode ser negativo
        else:
            return True #Regras do problema atingidas

    def objetivo(self):
        if((self.LadoDir[0]==3 and self.LadoDir[1]==3 ) and self.bote ==1):
            return True #Objetivo desejado
        else:
            return False #Ainda não atingiu o objetivo desejado

    def expEstados(self):
        operacoes = [[2,0],[0,2],[1,1],[0,1],[1,0]] #Matriz que armazena os movimentos possíveis
  
        bote = int(not(self.bote))#Define que, se o barco está em um lado, seu novo lado só pode ser o outro (0-esq) (1-dir)
        for i in range(5): #Gerar novos estados, realizando os movimentos armazenados na matriz "operacoes"
            if (bote==0):
                missionarioDir = self.LadoDir[0] - operacoes[i][0] 
                canibalDir = self.LadoDir[1] - operacoes[i][1] 
                missionarioEsq = self.LadoEsq[0] + operacoes[i][0] 
                canibalEsq = self.LadoEsq[1] + operacoes[i][1]
            else:
                missionarioDir = self.LadoDir[0] + operacoes[i][0]
                canibalDir = self.LadoDir[1] + operacoes[i][1]
                missionarioEsq = self.LadoEsq[0] - operacoes[i][0]
                canibalEsq = self.LadoEsq[1] - operacoes[i][1]

            filho = State([missionarioEsq,canibalEsq], [missionarioDir,canibalDir],bote) #Cria um objeto filho do tipo State
            filho.pai = self #O pai do objeto filho é o objeto atual
            if filho.verifica(): #Se o estado do objeto filho for válido
                self.filhos.append(filho) #Adicione ele na árvore

class Busca():
    def __init__(self,LadoEsq,LadoDir,bote):
        self.caminho = None #Não tem solução ainda
        self.arvore = [State(LadoEsq,LadoDir,bote)] #Inicializa árvore com um nó do tipo State

    def solucao(self):
        for no in self.arvore: #Itera sobre os elementos da árvore
            if(no.objetivo()): #Se o nó da árvore for o estado objetivo, faça
                self.caminho=[no] #Salve em caminho o nó atual
                while no.pai: #Agora salva o caminho reverso, enquanto o nó analisado tiver o pai diferente de no
                    self.caminho.insert(0,no.pai) #Armazene pai na cabeça da lista que contém a solução
                    no = no.pai #Nó agora é o pai do nó anterior
                break;
            else:
                no.expEstados() #O nó não é a solução, mas é válido, crie um novo nó na árvore
                self.arvore.extend(no.filhos) #Adicione este nó na árvore

class Imprime():
    def __init__(self,missionariosCanibais):
        self.missionariosCanibais = missionariosCanibais
    def imprime(self):
        self.missionariosCanibais.solucao()
        print("\t\tSOLUÇÃO:\n")
        i=0
        for estado in self.missionariosCanibais.caminho: #Itera sobre a lista caminho da Busca, que contém a solução do problema
              if(i!=0):
                print("\tEstado "+str(i+1)+":")
              else:print("\tEstado Inicial:")

              
              print("\t\t"+str(estado.LadoEsq[0]) + " missionários e " + str(estado.LadoEsq[1]) + " canibais à esquerda do rio.")
              print("\t\t"+str(estado.LadoDir[0]) + " misisonários e " + str(estado.LadoDir[1]) + " canibais à direita do rio.\n")
              i = i+1

class Principal():
    def main():
        missionariosCanibais = Busca([3,3],[0,0],0) #Configuração Inicial
        imprime = Imprime(missionariosCanibais)
        imprime.imprime()
        print("Explicação:\n"+"MC	MC\n33	00 ->Inicial\n31	02 ->Foi 2 ca"+"\n32	01 ->voltou 1 ca\n30	03 ->foi 2 ca\n31	02 ->voltou 1 ca"+"\n11	22 ->foi 2 mi"
        + "\n22	11 ->voltou 1 ca e 1 mi\n02	31 ->foi 2 mi\n03	30 ->voltou 1 ca\n01	32 ->foi 2 ca\n02	31 ->voltou 1 ca\n00	33 ->foi 2 ca"+"\n   FINAL!")
    if __name__ == '__main__':
        main()
```
